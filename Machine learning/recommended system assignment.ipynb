```python
# Converting long data into wide data using pivot table
books2=book.pivot_table(index='Userid',columns='Booktitle',values='Bookrating').reset_index(drop=True)
books2
```

```python
# Replacing the index values by unique UserId
books2.index=book['Userid'].unique()
books2
```

```python
# Impute those NaNs with 0 values
books2.fillna(0,inplace=True)
books2
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( books2.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = book.Userid.unique()
user_sim_df.columns = book.Userid.unique()
```

```python
user_sim_df.iloc[0:10, 0:10]
```

```python
# Nullifying diagonal values
np.fill_diagonal(user_sim,0)
user_sim
```

```python
def get_recommendation(customer_id):
        
        tem=list(user_sim_df.sort_values([customer_id],ascending=False).head(3).index)

        book_list=list(book[book['Userid'].isin(tem)]['Booktitle'])

        recommended_books = set(book_list)-set(book[book['Userid']==customer_id]['Booktitle'])
        
        return recommended_books
```
